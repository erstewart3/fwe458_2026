## What is Markdown

Markdown is a simple method for writing content on the web using plain textâ€”the same kind of text you already type every day. It relies on the regular alphabet along with a few common symbols, such as asterisks (*) and backticks (`), to add structure and emphasis.

Compared to bulky word-processing software, Markdown files are easy to share across computers, phones, and between different people. Because of this simplicity and portability, Markdown is rapidly becoming a standard tool for academics, scientists, writers, and others. Popular platforms like GitHub and Reddit use Markdown to format posts and comments.

Learning Markdown is straightforward and beginner-friendly. It doesnâ€™t support advanced styling like font colors, sizes, or typefaces. Instead, it focuses on basic presentation, allowing you to create headings, bold or emphasize text, and organize content with lists in a clean, consistent way.

## 1. Blocks and Structure
Markdown is **line-based**. Structure is determined before inline formatting.

### 1.1 Headings as Logical Structure
``` markdown
# 1. Introduction
## 2.1 Study Region
## 2.2 Data Sources
## 2.3 Model Description
```
# 1. Introduction
## 2.1 Study Region
## 2.2 Data Sources
## 2.3 Model Description

>Rules:
>- Headings represent logical hierarchy
>- Do not skip levels
>- Do not use headings for emphasis

### 1.2 Paragraphs
A paragraph is one or more consecutive non-blank lines.
```markdown
This is one paragraph.
Still the same paragraph.

This is a new paragraph.
```
This is one paragraph.
Still the same paragraph.  
This is a new paragraph.
> Blank lines create structure. Do not omit them.  
> **You can also use two spaces to break paragraphs**

### 1.3 Lists for Scientific Reasoning

Lists are for:

- Assumptions
- Steps
- Criteria
- Summary points

```markdown
We make the following assumptions:

- Steady-state conditions
- Homogeneous canopy structure
- Negligible advection
```
We make the following assumptions:

- Steady-state conditions
- Homogeneous canopy structure
- Negligible advection

```markdown
- Item
- Item
  - Subitem
```
- Item
- Item
  - Subitem

>**Indentation defines nesting, not alignment**

### 1.4 Blockquotes for Commentary

```markdown
> Note: Eddy covariance measurements are sensitive to footprint heterogeneity.
```
> Note: Eddy covariance measurements are sensitive to footprint heterogeneity.

Use blockquotes for:
- Notes
- Caveats
- Instructor comments
- Side explanations

## 2. Inline Grammar

### 2.1 Emphasis
```markdown
*italic*  
_italic_   
**bold**  
__bold__  
***bold italic***
_**bold italic**_
**_bold italic_**
```
*italic*  
_italic_  
**bold**  
__bold__  
***bold italic***  
_**bold italic**_  
**_bold italic_**

Guideline:

- Emphasis highlights meaning, not importance
- Use sparingly in scientific writing

### 2.2 Inline Code
Use inline code for:
- Variable names
- Functions
- File names

```markdown
We compute GPP using `compute_gpp()`.
```
Inline code is literal. No Markdown parsing inside.

### 2.3 Links
To create a link, enclose the link text in brackets (e.g., [UW-Madison]) and then follow it immediately with the URL in parentheses (e.g., (https://www.wisc.edu)).
```markdown
I love [UW-Madison](https://www.wisc.edu).
```
I love [UW-Madison](https://www.wisc.edu).

### 2.4 URLs and Email Addresses  
To quickly turn a URL or email address into a link, enclose it in angle brackets.
```markdown
<https://www.wisc.edu>
<min.chen@wisc.edu>
```
<https://www.wisc.edu>

<min.chen@wisc.edu>

### 2.5 Escaping Characters
To display a literal character that would otherwise be used to format text in a Markdown document, add a backslash (\) in front of the character.
```markdown
\* Without the backslash, this would be a bullet in an unordered list.
```
\* Without the backslash, this would be a bullet in an unordered list.

## 3. Markdown for Code Documentation

### 3.1 Why Code Documentation Matters in Environmental Data Science

Environmental data science code is rarely:
- Short
- Self-contained
- Used only once

Instead, it is often:
- Iterative
- Extended over years
- Reused by collaborators or future students
- Revisited after long gaps

Markdown is not a replacement for code comments.  
Markdown is where you explain **scientific intent**, **assumptions**, and **logic**.

> Code answers *how*.  
> Markdown answers *why*.

### 3.2 Separation of Concerns (Critical Principle)

A core principle of professional scientific computing is:

> **Do not mix implementation with explanation.**

- Code files explain *mechanics*
- Markdown explains *scientific reasoning*

This separation improves:
- Readability
- Reproducibility
- Long-term maintainability

## 4. Writing Mathematics in Markdown  

### 4.1 Why Complex Math Requires Better Writing (Not More Symbols)

Environmental data science routinely involves complex math equations. The difficulty is **not the math itself**, but **how it is communicated**.

### 4.2 Markdown + LaTeX: Division of Labor

- **Markdown** controls:
  - Structure
  - Narrative order
  - Where equations appear
- **LaTeX** controls:
  - Symbols
  - Alignment
  - Mathematical precision

Markdown decides **what role the equation plays**.  
LaTeX decides **how the equation is written**.

### 4.3 Inline Math: Referencing Symbols Without Breaking Flow

**Example 1: Radiative Transfer Variables**

Leaf reflectance ($\rho_l$) and transmittance ($\tau_l$) jointly determine canopy albedo.

**Example 2: Radiative Transfer Variables**

Net ecosystem exchange is defined as:

$$
\begin{aligned}
NEE &= F_{resp} - F_{assim} \\
    &= (R_{auto} + R_{hetero}) - A
\end{aligned}
$$

**Example 3: Heat Diffusion in Soil**

Soil temperature evolution is modeled using the heat diffusion equation:

$$
\frac{\partial T(z,t)}{\partial t}
=
\frac{\partial}{\partial z}
\left(
k(z) \frac{\partial T(z,t)}{\partial z}
\right)
$$

**Example 4: Areal Evapotranspiration**

Total evapotranspiration over a region $\Omega$ is given by:

$$
ET_{\Omega} =
\int_{\Omega} ET(x,y) \, dx \, dy
$$

**Example 5: Bayesian Inference**

We estimate model parameters $\theta$ using Bayes' theorem:

$$
p(\theta \mid y) =
\frac{p(y \mid \theta)\, p(\theta)}{p(y)}
$$
Here, $p(y \mid \theta)$ represents the likelihood of the observations given the parameters, while $p(\theta)$ encodes prior knowledge.

** Example 6: Soil Moisture Stress Function**

The stress function is defined as:

$$
f(\theta) =
\begin{cases}
0, & \theta \leq \theta_w \\
\frac{\theta - \theta_w}{\theta_f - \theta_w}, & \theta_w < \theta < \theta_f \\
1, & \theta \geq \theta_f
\end{cases}
$$

# 6. Table

Markdown tables are a popular feature in many Markdown applications, used to organize and present data clearly. They are created using pipes (|) to define columns and hyphens (-) to separate the header from the rest of the data.

**Basic Structure**

A Markdown table requires a header row, a separator line (using hyphens and optional colons), and at least one data row. The columns do not need to align perfectly in the raw source text for the rendered output to look correct, though alignment improves readability in the source.

```markdown

| Header 1 | Header 2 | Header 3 |
|---|---|---|
| Row 1 Col 1 | Row 1 Col 2 | Row 1 Col 3 |
| Row 2 Col 1 | Row 2 Col 2 | Row 2 Col 3 |
```

| Header 1 | Header 2 | Header 3 |
| --- | --- | --- |
| Row 1 Col 1 | Row 1 Col 2 | Row 1 Col 3 |
| Row 2 Col 1 | Row 2 Col 2 | Row 2 Col 3 |

**Text Alignment**

You can control the alignment of text within columns by adding colons (:) to the separator line.
```markdown
| Left-Aligned | Center-Aligned | Right-Aligned |
|:---|:---:|---:|
| Content | Content | Content |
| Content | Content | Content |
```

| Left-Aligned | Center-Aligned | Right-Aligned |
|:---|:---:|---:|
| Content | Content | Content |
| Content | Content | Content |

**Key Points**
- Pipes (|) separate columns within a row.
- Hyphens (-) form the mandatory separator line after the header. At least three hyphens are needed per column.
- Colons (:) are used within the separator line to indicate alignment (left, center, or right).
- Empty cells can be created by simply leaving the space between pipes blank.
- Tools like online Markdown Tables Generators can simplify the creation of complex tables.
